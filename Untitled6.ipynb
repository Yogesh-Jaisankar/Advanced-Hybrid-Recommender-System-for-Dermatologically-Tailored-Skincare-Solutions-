```python
import pandas as pd
import numpy as np
from scipy.linalg import eig

# Load the dataset
df = pd.read_csv('/Users/yogesh/Study/Supply_Chain_Management/cold_chain_logistics_data.csv')

# Step 1: Extract criteria importance columns for Cost, Delivery Time, and Environmental Impact
criteria_columns = ["Criteria Importance (Cost)", "Criteria Importance (Delivery Time)", "Criteria Importance (Environmental Impact)"]
criteria_data = df[criteria_columns]

# Step 2: Calculate the average importance for each criterion
avg_importance = criteria_data.mean()
cost, delivery_time, environmental_impact = avg_importance

# Step 3: Construct Pairwise Comparison Matrix based on averages
pairwise_matrix = np.array([
    [1, delivery_time / cost, environmental_impact / cost],
    [cost / delivery_time, 1, environmental_impact / delivery_time],
    [cost / environmental_impact, delivery_time / environmental_impact, 1]
])

print("Pairwise Comparison Matrix:\n", pairwise_matrix)

# Step 4: Calculate the AHP weights (eigenvector corresponding to the largest eigenvalue)
eigenvalues, eigenvectors = eig(pairwise_matrix)
max_index = np.argmax(eigenvalues)
weights = np.real(eigenvectors[:, max_index])
weights = weights / weights.sum()  # Normalize weights to sum to 1

```

#  Implementing NSGA-II

```python
import pandas as pd
import numpy as np
import random
from deap import base, creator, tools

# Load the data from the CSV file
data = pd.read_csv('/Users/yogesh/Study/Supply_Chain_Management/cold_chain_logistics_data.csv')

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from deap import base, creator, tools


# Extract the relevant columns for the optimization
transportation_costs = data['Total Transportation Cost'].values
delivery_times = data['Expected Delivery Time (Days)'].values
environmental_impacts = data['Total CO₂ Emissions (Kg)'].values   

# Set up the NSGA-II framework
creator.create("FitnessMin", base.Fitness, weights=(-1.0, -1.0, -1.0))  # Minimize costs, delivery time, and emissions
creator.create("Individual", list, fitness=creator.FitnessMin)

# Define the number of individuals in the population
population_size = 100

# Define the objective functions
def evaluate(individual):
    index = int(individual[0])  # Assuming the individual represents an index into the data
    cost = transportation_costs[index]
    delivery_time = delivery_times[index]
    emissions = environmental_impacts[index]
    
    return cost, delivery_time, emissions

# Register the evaluation and selection methods
toolbox = base.Toolbox()
toolbox.register("indices", np.random.permutation, len(transportation_costs))
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selNSGA2)

# Create the population
pop = toolbox.population(n=population_size)

# Run the optimization for a number of generations
num_generations = 50
for gen in range(num_generations):
    # Evaluate the individuals
    fitnesses = list(map(toolbox.evaluate, pop))
    for ind, fit in zip(pop, fitnesses):
        ind.fitness.values = fit

    # Select the next generation individuals
    offspring = toolbox.select(pop, len(pop))
    offspring = list(map(toolbox.clone, offspring))

    # Apply crossover and mutation
    for child1, child2 in zip(offspring[::2], offspring[1::2]):
        if np.random.rand() < 0.5:
            toolbox.mate(child1, child2)
            del child1.fitness.values
            del child2.fitness.values

    for mutant in offspring:
        if np.random.rand() < 0.2:
            toolbox.mutate(mutant)
            del mutant.fitness.values

    # Replace the old population by the offspring
    pop[:] = offspring

# Extract the results
fits = []
for ind in pop:
    if ind.fitness.valid:
        fits.append(ind.fitness.values)
    else:
        # Append a default tuple if fitness is not valid
        fits.append((float('inf'), float('inf'), float('inf')))

# Convert fits to a NumPy array
fits = np.array(fits)

# Filter out invalid fitness values for plotting
valid_fits = fits[~np.isinf(fits).any(axis=1)]

# Plot the results if there are valid fitness values
if valid_fits.size > 0:
    plt.scatter(valid_fits[:, 0], valid_fits[:, 1], c='blue', label='Pareto Front')
    plt.xlabel('Total Transportation Cost')
    plt.ylabel('Expected Delivery Time (Days)')
    plt.title('NSGA-II Pareto Front')
    plt.legend()
    plt.show()
else:
    print("No valid fitness values to plot.")
```

## Trade-off between cost and time: 

You can't improve both cost and delivery time simultaneously.

## Efficient solutions: 

Points on the Pareto Front are the best possible solutions given the trade-off.

## Decision making: 

Choose a solution based on your priorities (cost vs. time).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from deap import base, creator, tools



# Extract relevant columns for optimization
transportation_costs = data['Total Transportation Cost'].values
delivery_times = data['Expected Delivery Time (Days)'].values
environmental_impacts = data['Total CO₂ Emissions (Kg)'].values

# Set up NSGA-II framework
creator.create("FitnessMin", base.Fitness, weights=(-1.0, -1.0, -1.0))  # Minimize costs, time, and emissions
creator.create("Individual", list, fitness=creator.FitnessMin)

# Define population size and generations
population_size = 100
num_generations = 50

# Define brand-specific AHP weights
ahp_weights = {
    "Amul": {"cost": 0.5, "time": 0.3, "environment": 0.2},
    "Milky Mist": {"cost": 0.3, "time": 0.5, "environment": 0.2},
    "Kwality Walls": {"cost": 0.2, "time": 0.3, "environment": 0.5},
    "Arun": {"cost": 0.4, "time": 0.4, "environment": 0.2},
}

# Objective function considering AHP weights
def evaluate(individual, weights):
    index = int(individual[0])  # Assuming the individual represents an index into the data
    cost = transportation_costs[index] * weights["cost"]
    delivery_time = delivery_times[index] * weights["time"]
    emissions = environmental_impacts[index] * weights["environment"]
    
    return cost, delivery_time, emissions

# Register evaluation and selection methods
toolbox = base.Toolbox()
toolbox.register("indices", np.random.permutation, len(transportation_costs))
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selNSGA2)

# Run NSGA-II for each brand's preferences
results = {}
for brand, weights in ahp_weights.items():
    # Evaluate brand-specific population
    pop = toolbox.population(n=population_size)
    for gen in range(num_generations):
        # Evaluate individuals
        fitnesses = [evaluate(ind, weights) for ind in pop]
        for ind, fit in zip(pop, fitnesses):
            ind.fitness.values = fit

        # Selection and crossover
        offspring = toolbox.select(pop, len(pop))
        offspring = list(map(toolbox.clone, offspring))
        
        for child1, child2 in zip(offspring[::2], offspring[1::2]):
            if np.random.rand() < 0.5:
                toolbox.mate(child1, child2)
                del child1.fitness.values
                del child2.fitness.values

        # Mutation
        for mutant in offspring:
            if np.random.rand() < 0.2:
                toolbox.mutate(mutant)
                del mutant.fitness.values

        # Update population
        pop[:] = offspring
    
    # Collect and store final fitness values for the brand
    fits = np.array([ind.fitness.values for ind in pop if ind.fitness.valid])
    results[brand] = fits

# Plot Pareto fronts for each brand
plt.figure(figsize=(10, 7))
for brand, fits in results.items():
    plt.scatter(fits[:, 0], fits[:, 1], label=f'{brand} Pareto Front')
plt.xlabel('Weighted Transportation Cost')
plt.ylabel('Weighted Delivery Time')
plt.title('NSGA-II Pareto Front by Brand')
plt.legend()
plt.show()
```

```python
from tabulate import tabulate

# Display the top solutions for each brand in a tabular format
for brand, fits in results.items():
    # Create a list of dictionaries for the top solutions
    solutions = [{"Cost": solution[0], "Delivery Time": solution[1], "Emissions": solution[2]} for solution in fits[:5]]
    
    # Convert to table format and print
    print(f"\nTop solutions for {brand} (Weighted Cost, Delivery Time, Emissions):")
    print(tabulate(solutions, headers="keys", tablefmt="grid"))
```
